```python
import pandas as pd
file = 'like.xlsx'
xl = pd.ExcelFile(file)
df1 = xl.parse('Sheet1')
```

```python
df1.head(15)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array(list(range(20,85,5)))
x = x/100      
for i in range(20,95,5):
    n = i/100
    y = np.array(df1.loc[n,0.20:0.8])
    plt.plot(x, y, label=n)
    ax = plt.subplot(111)
    box = ax.get_position()
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12), fancybox=True, shadow=True, ncol=15)
plt.show()
```
